```python
import numpy as np
import matplotlib.pyplot as plt

# Define the camber line function
def camber_line(x_c, epsilon, c=1):
    z_c = np.zeros_like(x_c)
    for i, x in enumerate(x_c):
        if 0 <= x <= 0.25:
            z_c[i] = epsilon * c * (1 - 16 * (x - 0.25)**2)
        elif 0.25 < x <= 1:
            z_c[i] = epsilon * c * (1 - (16 / 9) * (x - 0.25)**2)
    return z_c / c  

# Generate x/c values
x_c = np.linspace(0, 1, 100)

# Define epsilon values
epsilons = [0.01, 0.03, 0.05]

# Plot camber lines
plt.figure(figsize=(12, 8))
for epsilon in epsilons:
    z_c = camber_line(x_c, epsilon)
    plt.plot(x_c, z_c, label=f' = {epsilon}')

plt.title("Camber Line for Different  Values")
plt.xlabel("x/c (Normalized Chordwise Position)")
plt.ylabel("z/c (Normalized Camber)")
plt.legend()
plt.grid(True)

plt.xlim(-0.1, 1.1)
plt.ylim(-0.02, 0.06) 

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the continuous function for alpha
def alpha_continuous(epsilon_values, C_L=1.13):
    # Simplified A0 calculation
    A0_cont = (-8 * epsilon_values / np.pi) * (np.pi / 3 - np.sqrt(3)) + (-8 * epsilon_values / (9 * np.pi)) * (2 * np.pi / 3 + np.sqrt(3))
    return C_L / (2 * np.pi) - A0_cont

# Generate a continuous range of epsilon values
epsilon_range = np.linspace(0, 0.05, 500)
alpha_values = alpha_continuous(epsilon_range)

# Plot alpha as a function of epsilon
plt.figure(figsize=(10, 6))
plt.plot(epsilon_range, alpha_values, label=r'$\alpha$ vs $\epsilon$', color='blue')
plt.title("Angle of Attack as a Function of Camber Parameter ()")
plt.xlabel("Epsilon () - Camber Parameter")
plt.ylabel("Angle of Attack (radians)")
plt.grid(True)
plt.legend()
plt.show()
```
